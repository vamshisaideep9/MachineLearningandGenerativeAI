Decision Tree Practical

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
df = sns.load_dataset('iris')
df.head(5)
```

```python
#Independent Features and dependent features

X = df.iloc[:,:-1]
y = iris.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#post pruning Technique

treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train, y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled=True)
```

Postpruning

```python
treemodel = DecisionTreeClassifier(max_depth=2)
```

```python
treemodel.fit(X_train, y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled=True)
```

```python
y_pred = treemodel.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report

score = accuracy_score(y_pred, y_test)
report = classification_report(y_pred, y_test)
```

```python
score
```

```python
report
```

Hyperparameter Tuning

```python
## Prepruning

parameters = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'splitter': ['best', 'random'],
    'max_depth': [1,2,3,4,5],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
cv = GridSearchCV(treemodel, param_grid=parameters, cv=5, scoring='accuracy') #cv=crossvalidation
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
cv.predict(X_train)
```

```python
treemodel = DecisionTreeClassifier(criterion='log_loss', max_depth=5, max_features='log2', splitter='random')
```

```python
treemodel.fit(X_train, y_train)
```

```python
from sklearn import tree

plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled=True)
```

```python
y_pred = treemodel.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, classification_report

score = accuracy_score(y_pred, y_test)
report = classification_report(y_pred, y_test)
```
